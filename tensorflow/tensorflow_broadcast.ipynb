```python
import tensorflow as tf
```

## 1.自动识别类

```python
a= tf.random.normal([4,35,9])
a.shape
```

```python
b = a+tf.random.normal([9])#不用调用broadcast它便会自动识别
#这里有一个自动扩张的过程，[9]-->[1,1,9]-->[4,35,9]，所以之和的shape依旧是[4,35,9]
#这里引入一个大维度和小维度的概念，自动右边对齐，默认大维度一致，后面加分，默认对所有班级的所有同学加分。
#注意1会被自动扩展，但是非1不会被扩展，并且类似加法，右对齐
b.shape
```

## 2调用函数类

```python
b = tf.broadcast_to(tf.random.normal([4,1,1,1]),[4,32,32,8])
b.shape
```

## 3.VS tf.tile

```python
a = tf.ones([3,4])
b = tf.broadcast_to(a,[2,3,4])#这里只用了[3,4]的内存便保存了[2,3,4]的内容，内存优化
b.shape
```

```python
a = tf.ones([3,4])
b = tf.expand_dims(a,axis=0)
b = tf.tile(b,[2,1,1])#占用了很多的内存，没有内存优化的作用
b.shape
```
