# 梯度，代表函数增大的方向

## 所以思考一下梯度更新的方式
## 当梯度是正数，即函数增大的方向，即减小它，往0的方向走
## 当梯度是负数，即函数减小的方向时，即增大它，往0的方向走，最终都会到达梯度为0的值


```python
w = tf.constant(1.)
x = tf.constant(2.)
with tf.GradientTape(persistent='True') as tape:
    #tape是一次性的算完就会释放，所以要想重复调用，设置persistent=’True‘，但是记得释放因为很占内存
    tape.watch([w])#跟踪参数的梯度,必须要，不然就会出现None的情况
    y = x*w  
```

```python
grad = tape.gradient(y,[w])#这里的梯度就是x，x=2，所以没错
grad
```
