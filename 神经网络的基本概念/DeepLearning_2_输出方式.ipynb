# 输出方式

```python
import tensorflow as tf
```

## 1.sigmoid将函数的值映射在[0,1]之间

```python
x = tf.linspace(-5.,5,5)#始末要求是浮点型不然会报错
x
```

```python
tf.sigmoid(x)
```

## 2.softmax将数据映射到[0,1]之间并且和为1

```python
a = tf.nn.softmax(x)#最多用来做概率预测的分类
```

```python
tf.reduce_sum(a)
```

```python
logit = tf.random.uniform([1,10],minval = 0,maxval=5)
logit
```

```python
prob = tf.nn.softmax(logit,axis=1)
prob
```

```python
tf.reduce_sum(prob)
```
